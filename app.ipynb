```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
```

```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
# Initial setup

def setup():
    driver = webdriver.Chrome()
    driver.get("https://techbrain.ai/")
    return driver

def teardown(driver):
    driver.quit()
```

```python
# Display the lesson list of ruby from home page

def list_ruby():
  driver=setup()
  list=driver.find_element(By.XPATH, "(//a/span[text()='Lists'])[1]")
  list.click()

  time.sleep(3)
  teardown(driver)
list_ruby()
```

```python
# Start course - ruby

def start_ruby():
  driver=setup()
  start=driver.find_element(By.XPATH, "(//a/span[text()='Start'])[1]")
  start.click()

  time.sleep(3)
  teardown(driver)
start_ruby()
```

```python
# Open and close lesson list of a course

def lesson():
  driver=setup()
  start=driver.find_element(By.XPATH, "(//a/span[text()='Start'])[1]")
  start.click()
  wait = WebDriverWait(driver, 60)
  # time.sleep(3)

  try:
    lessonlist=wait.until(
      EC.visibility_of_element_located((By.XPATH,  '//a[@data-modal-target="lessonListModal" or data-modal-toggle="lessonListModal"]'))
    )
    lessonlist.click()
    close=wait.until(
      EC.element_to_be_clickable((By.XPATH, "//span[text()='Close']"))
    )
    close.click()
    
  except Exception as e:
    print(e)
  finally:
    time.sleep(2) 
    teardown(driver)
  
lesson()
```

```python
# Automate lessons and quiz of course - ruby

def lesson():
  driver=setup()
  start=driver.find_element(By.XPATH, "(//a/span[text()='Start'])[1]")
  start.click()
  time.sleep(3)
  try:
    lesson_list=driver.find_element(By.XPATH, '//a[@data-modal-target="lessonListModal" or data-modal-toggle="lessonListModal"]')
    lesson_list.click()
    time.sleep(1)
    close=driver.find_element(By.XPATH, "//span[text()='Close']")
    close.click()
    time.sleep(2)
    page=1
    while (page<29):
      if (page==28):
        break
      else:
        try:
          nextLesson=driver.find_element(By.XPATH, "//button[@type='submit']")
          if nextLesson:
           nextLesson.click()
        except Exception:
          # nextLesson=driver.find_element(By.XPATH, "//button[@type='submit']")
          goquiz=driver.find_element(By.XPATH,"//span[text()='Go to Quiz']")
          if goquiz:
            goquiz.click()
            time.sleep(2)
            checkboxs=driver.find_elements(By.XPATH, "//input[@type='checkbox']")
            for checkbox in checkboxs:
              checkbox.click()
        time.sleep(2)
      page +=1
  except Exception as e:
    print(e)
  finally:
    time.sleep(3)
    teardown(driver)
lesson()
```
