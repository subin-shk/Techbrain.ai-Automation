```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
```

```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
# Initial setup

def setup():
    driver = webdriver.Chrome()
    driver.get("https://techbrain.ai/")
    return driver

def teardown(driver):
    driver.quit()
```

```python
# Display the lesson list of ruby from home page

def list_ruby():
  driver=setup()
  wait = WebDriverWait(driver, 60)
  list=driver.find_element(By.XPATH, "(//a/span[text()='Lists'])[1]")
  list.click()

  # time.sleep(3)
  wait.until(
    EC.visibility_of_element_located((By.XPATH, "//h1"))
  )
  teardown(driver)
list_ruby()
```

```python
# Start course - ruby

def start_ruby():
  driver=setup()
  start=driver.find_element(By.XPATH, "(//a/span[text()='Start'])[1]")
  start.click()

  time.sleep(3)
  teardown(driver)
start_ruby()
```

```python
# Open and close lesson list of a course

def lesson():
  driver=setup()
  start=driver.find_element(By.XPATH, "(//a/span[text()='Start'])[1]")
  start.click()
  wait = WebDriverWait(driver, 60)
  # time.sleep(3)

  try:
    lessonlist=wait.until(
      EC.visibility_of_element_located((By.XPATH,  '//a[@data-modal-target="lessonListModal" or data-modal-toggle="lessonListModal"]'))
    )
    lessonlist.click()
    close=wait.until(
      EC.element_to_be_clickable((By.XPATH, "//span[text()='Close']"))
    )
    close.click()
    
  except Exception as e:
    print(e)
  finally:
    time.sleep(2) 
    teardown(driver)
  
lesson()
```

```python
def login(driver):
  # driver=setup()
  wait = WebDriverWait(driver, 60)
  signin=driver.find_element(By.XPATH, "(//a/span[text()='Sign in'])[1]")
  signin.click()
  wait.until(
    EC.visibility_of_element_located((By.CSS_SELECTOR, "input#user_email"))
  )
  try:
    email=driver.find_element(By.CSS_SELECTOR, "input#user_email")
    email.send_keys('aatlzg@endtest-mail.io')
    # email.send_keys(random_mail)
    password=driver.find_element(By.CSS_SELECTOR, "input#user_password")
    password.send_keys('P@ssword123')
    login=driver.find_element(By.NAME, 'commit')
    login.click()
  except Exception as e:
    print(e)
  finally:
    wait.until(
      EC.visibility_of_element_located((By.XPATH, "//h1[text()='Courses']"))
    )
  siginsuccess=wait.until(
    EC.visibility_of_element_located((By.XPATH, "(//a/span[text()='Sign out'])[1]"))
    )
  
  actualResult=siginsuccess.text
  expectedResult="Sign out"
  try:
    assert actualResult == expectedResult, "Login unsuccessful"
    print("Login Successful")
  except Exception as e:
    print(e)
  # teardown(driver)
  return driver
login(setup())
```

```python
from selenium.webdriver.common.alert import Alert

def signout():
  driver=setup()
  wait = WebDriverWait(driver, 60)
  login(driver)
  driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
  driver.find_element(By.XPATH,"//span[text()='Sign out']").click()

  alert = Alert(driver)
  print(alert.text)
  alert.accept()

  logoutsuccess=wait.until(
    EC.visibility_of_element_located((By.XPATH, "(//a/span[text()='Sign in'])[1]"))
    )
  time.sleep(3)
  actualResult=logoutsuccess.text
  expectedResult="Sign in"
  try:
    assert actualResult == expectedResult, "Logout unsuccessful"
    print("Logout Successful")
  except Exception as e:
    print(e)
  teardown(driver)
signout()
```

```python
# Automate lessons and quiz of course - ruby

def lesson():
  driver=setup()
  wait = WebDriverWait(driver, 60)
  start=driver.find_element(By.XPATH, "(//a/span[text()='Start'])[1]")
  start.click()
  time.sleep(3)
  
  try:
    page=1
    while (page<29):
      if (page==28):
        break
      else:
        try:
          nextLesson=driver.find_element(By.XPATH, "//button[@type='submit']")
          if nextLesson:
           nextLesson.click()
        except Exception:
          # nextLesson=driver.find_element(By.XPATH, "//button[@type='submit']")
          goquiz=driver.find_element(By.XPATH,"//span[text()='Go to Quiz']")
          if goquiz:
            goquiz.click()
            # time.sleep(2)
            wait.until(
              EC.visibility_of_element_located((By.XPATH, "//h2[contains(text(),'Quiz')]"))
            )
            checkboxs=driver.find_elements(By.XPATH, "//input[@type='checkbox']")
            for checkbox in checkboxs:
              checkbox.click()
        time.sleep(2)
      page +=1
  except Exception as e:
    print(e)
  finally:
    time.sleep(3)
    teardown(driver)
lesson()
```
